#### This notebook will be used to learn best practices and how to clean up data.

```python
import pandas as pd
import numpy as np
```

in notebooks, you can print data just by writing the df name or the numpy array name.

```python
df = pd.read_excel('./assets/Customer_Call_List.xlsx')
df
```

the very first thing you're usually gonna wanna do, is get rid of duplicates.

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns=['Not_Useful_Column'])
df
```

```python
#we have either lstrip, or rstrip (left or right), strip takes both sides, and removes the value we pass.
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df
```

while we can use lstrip and rstrip individually, we can also combine them.

```python
df['Last_Name'] = df["Last_Name"].str.strip("123._/")
df
#123 removes numbers on any side, . removes any periods on any side, _ removes itself, and so does /. any character you enter here gets removed.
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
#this is an example of a regex (regular expression), what it's doing is saying that replace all these characters, that is not a-z, A-Z, 0-9, with nothing.
df
```

```python
#convert phone number to string first.
df["Phone_Number"].apply(lambda x: str(x)) #this creates a for loop that goes through every digit and converts it to string.
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #now split into 3 digit arrays, and add - in between.
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '') #since null values converted to string, they say nan-- or sth else now, so we replace with nothing.
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
df[["Street", "City", "Zip"]] = df["Address"].str.split(',', expand=True) 
#splits everything based on comma, and expand means it's assigning that split value to a new column.
df
```

```python
del df["Address"]
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a', '') #replaces any non null strings showing nan with nothing.
df = df.replace('NaN', '')
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

#we could have used df.dropna(subset = "Phone_Number") and that way it would have dropped the cols with null values for phone numbers.
df
```

```python
del df["Zip"]
df
```

```python
df = df.reset_index(drop=True) #resets the index, and preventes the current indexing from being added as a row.
df
```

#### This concludes the cleanup for this data.

- remember to drop duplicates
- null values that aren't relevant
- split into relevant cols when possible
- drop cols that aren't relevant
- and finally, save the new data

```python
df.to_excel('./assets/Customer_Call_List_Edited.xlsx')
```
