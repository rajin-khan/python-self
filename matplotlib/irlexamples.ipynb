## We will be taking real data and visualizing them in this notebook.

#### First, the libraries needed:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line graph example:

```python
gas = pd.read_csv('./assets/gas_prices.csv')
gas #verifying the import
```

```python
plt.figure(figsize=(8, 5))

plt.title('Gas Prices over time', fontdict={'fontweight' : 'bold'})

plt.plot(gas.Year, gas.USA, 'blue', marker='.', label='USA') #on the x axis we have year, y axis gets the price for each country.
plt.plot(gas.Year, gas.Canada, 'orange', marker='.', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'red', marker='.', label='South Korea') #both formats are fine, but the [] format helps for cols with spaced names
plt.plot(gas.Year, gas.Japan, 'pink', marker='.', label='Japan')

#you could alternatively loop through the data:
#for country in gas:
#    if country != 'Year':
#        plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::2]) #slice operator, picks from the start of the year col to the end, iterating by 2.
#or we could do: plt.xticks(np.arange(1990, 2009, 2))
plt.xlabel('Years')
plt.ylabel('Price in $')

plt.legend() #you can move the legend outside the graph too, look it up later
plt.tight_layout()
plt.savefig('./generated/gaspricegraph.png', dpi=500)
plt.show()
```

### Let's work with the **FIFA20** data now.

```python
fifa = pd.read_csv('./assets/fifa_data.csv')
fifa.head()
```

### Histogram example:

```python
plt.hist(fifa.Overall)

plt.show()
```

### Formatting the Histogram a bit more:

```python
plt.figure(figsize=(6, 8))

plt.title('Distribution of ratings in FIFA20', fontdict={'fontweight': 'bold'})

bins = np.arange(40, 110, 10) #creates a list from 0-100, intervals of 10

plt.hist(fifa.Overall, color='orange', bins=bins) #bins are the ranges we want, in a list.

plt.xlabel('Ratings')
plt.ylabel('Number of players')
plt.xticks(bins) #making sure the x axis ticks correspond to the ranges.

plt.tight_layout()
plt.savefig('./generated/fifahist.png')
plt.show()
```

### Pie charts:

```python
#figuring out what we may wanna use for the pie chart
fifa['Preferred Foot']
```

#### Let's go with % of preferred foot:

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].shape[0]
#counts could be used, but it gives a warning, so we just take the shape of the array where the value is left, which is basically the count.
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].shape[0]
```

#### Finally, let's plot:

```python
plt.title('Foot preference of players in FIFA20', fontdict={'fontweight':'bold'})

labels = ['Left', 'Right'] #a list of labels corresponding to the input list of numbers
#could pass in a list of correspoinding colours too: color = ['orange', 'blue'], then assign colors=color in the plot function.

plt.pie([left, right], labels=labels, autopct='%.2f %%') #two %% sign is like an escape sequence for the % sign.
#takes a list of numbers first.
#then a list of the labels
#maybe show the percentage too

plt.tight_layout
plt.savefig('./generated/footprefpi.png', dpi=500)
plt.show()
```
