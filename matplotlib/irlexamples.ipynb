## We will be taking real data and visualizing them in this notebook.

#### First, the libraries needed:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line graph example:

```python
gas = pd.read_csv('./assets/gas_prices.csv')
gas #verifying the import
```

```python
plt.figure(figsize=(8, 5))

plt.title('Gas Prices over time', fontdict={'fontweight' : 'bold'})

plt.plot(gas.Year, gas.USA, 'blue', marker='.', label='USA') #on the x axis we have year, y axis gets the price for each country.
plt.plot(gas.Year, gas.Canada, 'orange', marker='.', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'red', marker='.', label='South Korea') #both formats are fine, but the [] format helps for cols with spaced names
plt.plot(gas.Year, gas.Japan, 'pink', marker='.', label='Japan')

#you could alternatively loop through the data:
#for country in gas:
#    if country != 'Year':
#        plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::2]) #slice operator, picks from the start of the year col to the end, iterating by 2.
#or we could do: plt.xticks(np.arange(1990, 2009, 2))
plt.xlabel('Years')
plt.ylabel('Price in $')

plt.legend() #you can move the legend outside the graph too, look it up later
plt.tight_layout()
plt.savefig('./generated/gaspricegraph.png', dpi=500)
plt.show()
```
