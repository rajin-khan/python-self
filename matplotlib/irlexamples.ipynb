## We will be taking real data and visualizing them in this notebook.

#### First, the libraries needed:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line graph example:

```python
gas = pd.read_csv('./assets/gas_prices.csv')
gas #verifying the import
```

```python
plt.figure(figsize=(8, 5))

plt.title('Gas Prices over time', fontdict={'fontweight' : 'bold'})

plt.plot(gas.Year, gas.USA, 'blue', marker='.', label='USA') #on the x axis we have year, y axis gets the price for each country.
plt.plot(gas.Year, gas.Canada, 'orange', marker='.', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'red', marker='.', label='South Korea') #both formats are fine, but the [] format helps for cols with spaced names
plt.plot(gas.Year, gas.Japan, 'pink', marker='.', label='Japan')

#you could alternatively loop through the data:
#for country in gas:
#    if country != 'Year':
#        plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::2]) #slice operator, picks from the start of the year col to the end, iterating by 2.
#or we could do: plt.xticks(np.arange(1990, 2009, 2))
plt.xlabel('Years')
plt.ylabel('Price in $')

plt.legend() #you can move the legend outside the graph too, look it up later
plt.tight_layout()
plt.savefig('./generated/gaspricegraph.png', dpi=500)
plt.show()
```

### Let's work with the **FIFA20** data now.

```python
fifa = pd.read_csv('./assets/FIFA23_official_data.csv')
fifa.head()
```

### Histogram example:

```python
plt.hist(fifa.Overall)

plt.show()
```

### Formatting the Histogram a bit more:

```python
plt.figure(figsize=(6, 8))

plt.title('Distribution of ratings in FIFA23', fontdict={'fontweight': 'bold'})

bins = np.arange(40, 110, 10) #creates a list from 0-100, intervals of 10

plt.hist(fifa.Overall, color='orange', bins=bins) #bins are the ranges we want, in a list.

plt.xlabel('Ratings')
plt.ylabel('Number of players')
plt.xticks(bins) #making sure the x axis ticks correspond to the ranges.

plt.tight_layout()
plt.savefig('./generated/fifahist.png')
plt.show()
```

### Pie charts:

```python
#figuring out what we may wanna use for the pie chart
fifa['Preferred Foot']
```

#### Let's go with % of preferred foot:

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].shape[0]
#counts could be used, but it gives a warning, so we just take the shape of the array where the value is left, which is basically the count.
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].shape[0]
```

#### Finally, let's plot:

```python
plt.title('Foot preference of players in FIFA23', fontdict={'fontweight':'bold'})

labels = ['Left', 'Right'] #a list of labels corresponding to the input list of numbers
#could pass in a list of correspoinding colours too: color = ['orange', 'blue'], then assign colors=color in the plot function.

plt.pie([left, right], labels=labels, autopct='%.2f %%') #two %% sign is like an escape sequence for the % sign.
#takes a list of numbers first.
#then a list of the labels
#maybe show the percentage too

plt.tight_layout
plt.savefig('./generated/footprefpi.png', dpi=500)
plt.show()
```

#### Let's look at the weight of all the players for another pie chart:

```python
fifa['Weight'] = [int(x.strip('kg')) if type(x)==str else x for x in fifa.Weight] 
#if the value exists, strip lbs off the end, if null, leave as is, in a for loop, then convert it to an int.

#creating the weight categories:
light = fifa.loc[fifa.Weight<60].shape[0] #counting the cols where weight < 125 pounds
light_medium = fifa.loc[(fifa.Weight>=60) & (fifa.Weight<70)].shape[0]
medium = fifa.loc[(fifa.Weight>=70) & (fifa.Weight<80)].shape[0]
medium_heavy = fifa.loc[(fifa.Weight>=80) & (fifa.Weight<90)].shape[0]
heavy = fifa.loc[fifa.Weight>=90].shape[0]

light, light_medium, medium, medium_heavy, heavy
```

```python
plt.title('Weights of athletes in FIFA23 (in kg)', fontdict={'fontweight':'bold'})

#instead of manually passing a color scheme, like this:
#colors = ['black', 'red', 'orange', 'green', 'blue']
#better to use a stylesheet, as found here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
#to use, simply:
plt.style.use('ggplot')

#putting the weights in an array:
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 60', '60-70', '70-80', '80-90', 'Over 90']
explode = (.4, .1, 0, 0, .4) #corresponds to each section of the label and how much we want it to be exploded/separated 0 being no separation, 1 being the most.

plt.pie(weights, labels=labels, labeldistance=1.2, autopct='%.2f %%', pctdistance=0.8, explode=explode)
#pctdistance determines the placement of the %s in the chart.
#explode takes an array of values corresponding to the size of the label.

plt.tight_layout()
plt.savefig('./generated/weights.png', dpi=500)
plt.show()
```

### How might we compare different teams?
#### A viable way may be a box and whiskers plot (as it shows a player's max rating, min, and mean).

```python
fifa.head()
```

### Now, let's plot a Box & Whiskers example:

```python
#let's single out two teams and store them in separate dfs
real = fifa.loc[fifa.Club == 'Real Madrid CF']['Overall'] #we take just the overall player values
barca = fifa.loc[fifa.Club == 'FC Barcelona']['Overall'] 
manu = fifa.loc[fifa.Club == 'Manchester United']['Overall'] 

#labels for the box and whiskers plot
labels = ['Real Madrid', 'FC Barcelona', 'Manchester United']

#now, we plot:
plt.style.use('default') #setting it back in this cell as putting it like that in one cell results in changing the theme everywhere.

plt.figure(figsize=(5, 8))

plt.title('Team comparison in FIFA23', fontdict={'fontweight':'bold'})

plt.boxplot([real, barca, manu], tick_labels=labels)
#for box plot, labels is deprecated, we use tick_labels

plt.ylabel('FIFA Overall Rating')

plt.tight_layout()
plt.savefig('./generated/teamcomp.png', dpi=500)
plt.show()
```
