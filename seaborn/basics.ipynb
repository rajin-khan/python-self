## We will learn data visualization using Seaborn in this one notebook.
#### Covered concepts will include basic graphs and other visualization techniques. Seaborn is much easier to use as it assumes a lot of things when we give it info. This leads to us being able to generate plots, often with only one line of code.
**(everything covered can of course be found in the documentation). Documentation linked [here](https://seaborn.pydata.org/tutorial.html).**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid', palette='deep')
```

### Seaborn has built in datasets. 
#### We will be utilizing them in this tutorial. Of course, we could import into dataframes using pandas, but we're doing this to make things easier when learning.

```python
#may be needed when importing datasets from seaborn's available library.
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
#you do not need to know how this works.

crash_df = sns.load_dataset('car_crashes')
crash_df.head(10)
```

### Let's start off with Distribution Plots.

##### Distribution Plot:

```python
sns.displot(crash_df['not_distracted'], bins=25)
#bins creates the distribution bucket
plt.savefig('generated/distributionplot.png', dpi=500)
```

#### Joint Plot: (compares two distributions)

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
#here we have what goes on the x axis, and the y axis. then the data part takes in the dataframe. kind takes in what kind of line do we want to fit.
# 'reg' means based on regression.
plt.savefig('generated/jointplotregression.png', dpi=500)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
#kde = kernel density estimation
plt.savefig('generated/jointplotkde.png', dpi=500)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hist')
plt.savefig('generated/jointplothist.png', dpi=500)
```

#### KDE Plot: (Kernel Density Estimation)

```python
sns.kdeplot(crash_df['alcohol'])
plt.savefig('generated/kdeplot.png', dpi=500)
```

#### Pair Plots:

```python
sns.pairplot(crash_df) #generates all possible combinations of plots.
plt.savefig('generated/allpairplotcrashes.png', dpi=500)
```

### Let's load a different dataset to try out other things too.

```python
tips_df = sns.load_dataset('tips')
tips_df.head(10)
```

```python
sns.pairplot(tips_df, hue='smoker', palette='magma')
#hue decides what feature we want to base he differences on
#you can find more palettes here: https://matplotlib.org/stable/users/explain/colors/colormaps.html
plt.savefig('generated/pairplottips.png', dpi=500)
```

###  Rug Plots:

```python
sns.rugplot(tips_df['tip'])
#denser where more common values
plt.savefig('generated/rugplottips.png', dpi=500)
```

### Styling

```python
sns.set_style('white')

sns.set_context('paper', font_scale=1.4) #for context, 'paper' is smaller, for jupyter notebooks, 'talk' is for presentations, 'poster' is for slides, the text and axes just get bigger.

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=True) #removes the axes lines.

plt.savefig('generated/jointplotcrashes.png', dpi=500)
```
